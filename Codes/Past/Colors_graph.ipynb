```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracin de los datos
t = np.linspace(0, 2 * np.pi, 100)  # Tiempo de 0 a 2 con 100 puntos

# Generacin de tres senoidales con datos aleatorios
amplitudes = np.random.rand(3) * 2  # Amplitud aleatoria entre 0 y 2
frequencies = np.random.rand(3) * 2  # Frecuencia aleatoria entre 0 y 2
phases = np.random.rand(3) * 2 * np.pi  # Fase aleatoria entre 0 y 2

# Calculo de las tres senoidales
y1 = [np.random.random() for _ in range(len(t))]#amplitudes[0] * np.sin(frequencies[0] * t + phases[0])
y2 = [np.random.random() for _ in range(len(t))]#amplitudes[1] * np.sin(frequencies[1] * t + phases[1])
y3 = [np.random.random() for _ in range(len(t))]#amplitudes[2] * np.sin(frequencies[2] * t + phases[2])

# Graficar
plt.figure(figsize=(10, 6), facecolor='lightgrey')
plt.plot(t, y1, label='Senoidal 1', color='orangered', linewidth = 1.2, linestyle="-", alpha=0.5)
plt.plot(t, y2, label='Senoidal 2', color='limegreen', linewidth = 1.2, linestyle="--")
plt.plot(t, y3, label='Senoidal 3', color='navy', linewidth = 1, linestyle="-.")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Tres ondas senoidales con datos aleatorios")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.signal import savgol_filter
```

```python
# Cumulative distribution function
#ecdf = ECDF(x=col)
plt.figure(figsize =(16,8), tight_layout=True)
#plt.figure(figsize=(10, 6), facecolor='lightgrey')
#gs = gridspec.GridSpec(2, 6)
#ax = fig.add_subplot(gs[1, 3:6])
#plt.plot(t, t, color="#515ea1")
plt.plot(t, y1, label='Senoidal 1', color='limegreen', linewidth = 2, linestyle="-.") #Raw
plt.plot(t, y3, label='Senoidal 3', color='navy', linewidth = 1, linestyle="-") #Filter
plt.plot(t, y2, label='Senoidal 2', color='orangered', linewidth = 1.2, linestyle="-") #ESA calibrated
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Tres ondas senoidales con datos aleatorios")
plt.legend()
plt.grid(True)
#ax.set_ylabel("CDF")
#ax.set_xlabel()
plt.tight_layout()
plt.show()
```

```python
w = 15
p = 3
y1s = savgol_filter(y1, window_length=w, polyorder=p)
y2s = savgol_filter(y2, window_length=w, polyorder=p)
y3s = savgol_filter(y3, window_length=w, polyorder=p)
# Cumulative distribution function
#ecdf = ECDF(x=col)
plt.figure(figsize =(16,8), tight_layout=True)
#plt.figure(figsize=(10, 6), facecolor='lightgrey')
#gs = gridspec.GridSpec(2, 6)
#ax = fig.add_subplot(gs[1, 3:6])
#plt.plot(t, t, color="#515ea1")
plt.plot(t, y1s, label='Senoidal 1', color='limegreen', linewidth = 2, linestyle="-.") #Raw
plt.plot(t, y3s, label='Senoidal 3', color='navy', linewidth = 1, linestyle="-") #Filter
plt.plot(t, y2s, label='Senoidal 2', color='orangered', linewidth = 1.2, linestyle="-") #ESA calibrated
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Tres ondas senoidales con datos aleatorios")
plt.legend()
plt.grid(True)
#ax.set_ylabel("CDF")
#ax.set_xlabel()
plt.tight_layout()
plt.show()
```
